**퍼셉트론 이란?**  

퍼셉트론은 다수의 신호를 입력받아 하나의 신호를 출력한다. 퍼셉트론 신호도 흐름을 만들고 정보를 앞으로 전달한다. 신호는 '흐른다/안흐른다'(1/0) 두가지 값을 가질 수 있다.  

입력신호가 뉴런에 보내질 때는 각각 고유한 가중치가 곱해진다. 뉴런에서 보내온 신호의 총합이 정해진 한계를 넘어설 때만 1을 출력한다. 그 한계값을 임계값이라고도 한다. 

```python
def AND(x1,x2):
    w1,w2,theta=0.5,0.5,0.7
    tmp=x1*w1+x2*w2
    if tmp <= theta:
        return 0
    elif tmp > theta:
        return 1
    
```

```python
AND(0,0)
```

```python
AND(1,0)
```

```python
AND(0,1)
```

```python
AND(1,1)
```

```python
import numpy as np
x= np.array([0,1])
w=np.array([0.5,0.5])
b=-0.7
w*x
```

```python
np.sum(w*x)+b
```

```python
def AND(x1,x2):
    x=np.array([x1,x2])
    w=np.array([0.5,0.5])
    b=-0.7
    tmp = np.sum(w*x)+b
    if tmp<=0:
        return 0
    else:
        return 1
```

w1,w2을 가중치로 b를 편향으로 서로 구별하기도 하지만 , 문맥에 따라 모두 가중치라고 할 때도 있다.   

퍼셉트론은 직선 하나로 나눈 영역만 표현할 수 있다는 한계가 있다
